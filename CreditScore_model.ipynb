```python


# 1) Imports
import os
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split, StratifiedKFold, cross_validate
from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline
from imblearn.pipeline import Pipeline as ImbPipeline
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import OneHotEncoder, StandardScaler, OrdinalEncoder
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from imblearn.over_sampling import SMOTE
from sklearn.metrics import (classification_report, confusion_matrix,
                             roc_auc_score, precision_recall_fscore_support,
                             roc_curve, precision_recall_curve, average_precision_score)
import joblib
import warnings
warnings.filterwarnings('ignore')
sns.set(style="whitegrid")

# 2) Download & load data (German credit - UCI)
!wget -q https://archive.ics.uci.edu/ml/machine-learning-databases/statlog/german/german.data -O german_credit.csv

cols = ["Status_of_existing_checking_account", "Duration_in_month", "Credit_history",
        "Purpose", "Credit_amount", "Savings_account", "Employment_since", "Installment_rate",
        "Personal_status_and_sex", "Other_debtors", "Residence_since", "Property",
        "Age", "Other_installment_plans", "Housing", "Number_of_existing_credits",
        "Job", "Number_of_dependents", "Telephone", "Foreign_worker", "target"]

df = pd.read_csv("german_credit.csv", sep=" ", header=None, names=cols)
# map target: 1=good -> 0, 2=bad -> 1 (so 1 means bad)
df["target"] = df["target"].map({1: 0, 2: 1})

print("Shape:", df.shape)
print("Target distribution:\n", df["target"].value_counts(normalize=True))
display(df.head())

# 3) Quick EDA (print numeric summary & some value counts)
print("\nNumeric summary:")
display(df.describe().T)

print("\nSample categorical value counts (first 6 categorical columns):")
for c in df.select_dtypes(include=['object']).columns[:6]:
    print(f"\n-- {c} --")
    print(df[c].value_counts().to_dict())

# 4) Feature engineering (domain-driven)
# The German dataset lacks explicit income; we'll create useful derived features:
# - credit_per_month = Credit_amount / Duration_in_month
# - high_installment = Installment_rate > median
# - young = Age < 30, senior = Age > 60
# - num_existing_credits (already present) -> cap outliers
# - encode some categorical groups later

df = df.copy()
df["credit_per_month"] = df["Credit_amount"] / (df["Duration_in_month"].replace(0,1))
df["credit_per_month"] = df["credit_per_month"].round(2)
df["high_installment"] = (df["Installment_rate"] > df["Installment_rate"].median()).astype(int)
df["age_group"] = pd.cut(df["Age"], bins=[20,30,45,60,100], labels=["20-30","31-45","46-60","60+"])
# cap number_of_existing_credits
df["Number_of_existing_credits"] = df["Number_of_existing_credits"].clip(upper=5)
# flag for many dependents
df["many_dependents"] = (df["Number_of_dependents"] > 1).astype(int)

# 5) Prepare feature lists (numerical and categorical)
# Choose a compact set of features for modeling (numeric + engineered + some categorical)
numeric_features = [
    "Duration_in_month", "Credit_amount", "Installment_rate", "Age",
    "credit_per_month"
]
engineered_numeric = ["Number_of_existing_credits"]
numeric_features += engineered_numeric

# choose categorical columns (original + engineered)
categorical_features = [
    "Status_of_existing_checking_account", "Credit_history", "Purpose",
    "Savings_account", "Employment_since", "Personal_status_and_sex",
    "Other_debtors", "Property", "Other_installment_plans", "Housing",
    "Job", "Telephone", "Foreign_worker", "age_group", "high_installment", "many_dependents"
]
# ensure columns exist
categorical_features = [c for c in categorical_features if c in df.columns]

print("\nNumeric features:", numeric_features)
print("Categorical features (sample):", categorical_features[:8])

# 6) Train-test split
target_col = "target"
X = df[numeric_features + categorical_features]
y = df[target_col]
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.25, stratify=y, random_state=42)

print("\nTrain shape:", X_train.shape, "Test shape:", X_test.shape)

# 7) Preprocessing pipelines
numeric_transformer = Pipeline(steps=[
    ("imputer", SimpleImputer(strategy="median")),
    ("scaler", StandardScaler())
])

# For many small cardinality categoricals, OneHotEncoder is OK; for larger, Ordinal might be better.
categorical_transformer = Pipeline(steps=[
    ("imputer", SimpleImputer(strategy="most_frequent")),
    ("onehot", OneHotEncoder(handle_unknown="ignore"))
])

preprocessor = ColumnTransformer(transformers=[
    ("num", numeric_transformer, numeric_features),
    ("cat", categorical_transformer, categorical_features)
])

# 8) Model definitions
models = {
    "LogisticRegression": LogisticRegression(max_iter=1000),
    "DecisionTree": DecisionTreeClassifier(random_state=42),
    "RandomForest": RandomForestClassifier(n_estimators=300, random_state=42, n_jobs=-1)
}

# 9) Cross-validate and choose best model (using pipeline with SMOTE inside training pipeline)
cv = StratifiedKFold(n_splits=5, shuffle=True, random_state=42)
scoring = {"roc_auc": "roc_auc", "precision": "precision", "recall": "recall", "f1": "f1"}

cv_results = {}
for name, clf in models.items():
    pipe = ImbPipeline(steps=[
        ("preproc", preprocessor),
        ("smote", SMOTE(random_state=42)),
        ("clf", clf)
    ])
    scores = cross_validate(pipe, X_train, y_train, cv=cv, scoring=scoring, n_jobs=-1)
    cv_results[name] = {metric: np.mean(scores[f"test_{metric}"]) for metric in scoring}

cv_df = pd.DataFrame(cv_results).T
cv_df = cv_df.sort_values("roc_auc", ascending=False)
print("\nCross-validation results (mean):")
display(cv_df)

best_model_name = cv_df.index[0]
print("\nBest model by ROC-AUC:", best_model_name)

# 10) Retrain chosen model on full train set
best_clf = models[best_model_name]
final_pipeline = ImbPipeline(steps=[
    ("preproc", preprocessor),
    ("smote", SMOTE(random_state=42)),
    ("clf", best_clf)
])
final_pipeline.fit(X_train, y_train)

# 11) Evaluate on test set (multiple metrics)
y_proba = final_pipeline.predict_proba(X_test)[:,1]
y_pred = final_pipeline.predict(X_test)

roc_auc = roc_auc_score(y_test, y_proba)
precision, recall, f1, _ = precision_recall_fscore_support(y_test, y_pred, average='binary', zero_division=0)
ap = average_precision_score(y_test, y_proba)
cm = confusion_matrix(y_test, y_pred)

print(f"\nTest ROC-AUC: {roc_auc:.4f}")
print(f"Average Precision (PR AUC): {ap:.4f}")
print(f"Precision: {precision:.4f} | Recall: {recall:.4f} | F1: {f1:.4f}")
print("\nConfusion Matrix:\n", cm)
print("\nClassification report:\n", classification_report(y_test, y_pred, digits=4))

# 12) Plots: ROC and Precision-Recall
fpr, tpr, roc_th = roc_curve(y_test, y_proba)
prec, rec, pr_th = precision_recall_curve(y_test, y_proba)

plt.figure(figsize=(6,5))
plt.plot(fpr, tpr, label=f"ROC AUC={roc_auc:.3f}")
plt.plot([0,1],[0,1],"k--")
plt.xlabel("False Positive Rate"); plt.ylabel("True Positive Rate")
plt.title("ROC Curve")
plt.legend()
plt.show()

plt.figure(figsize=(6,5))
plt.plot(rec, prec, label=f"AP={ap:.3f}")
plt.xlabel("Recall"); plt.ylabel("Precision")
plt.title("Precision-Recall Curve")
plt.legend()
plt.show()

# 13) Feature importance (if tree)
if best_model_name in ["DecisionTree", "RandomForest"]:
    # recover transformed feature names
    ohe = final_pipeline.named_steps["preproc"].named_transformers_["cat"].named_steps["onehot"]
    cat_names = list(ohe.get_feature_names_out(categorical_features))
    feat_names = numeric_features + cat_names
    importances = final_pipeline.named_steps["clf"].feature_importances_
    fi = pd.Series(importances, index=feat_names).sort_values(ascending=False).head(20)
    print("\nTop feature importances:")
    display(fi)
    plt.figure(figsize=(7,6))
    sns.barplot(x=fi.values, y=fi.index)
    plt.title("Top Features")
    plt.show()

# 14) Save trained inference pipeline (preprocessor + classifier, WITHOUT SMOTE)
# Build a lightweight inference pipeline (no SMOTE)
from sklearn.pipeline import Pipeline as SkPipeline
inference_pipeline = SkPipeline(steps=[
    ("preproc", final_pipeline.named_steps["preproc"]),
    ("clf", final_pipeline.named_steps["clf"])
])
os.makedirs("models", exist_ok=True)
joblib.dump(inference_pipeline, "models/credit_inference_pipeline.pkl")
print("\nSaved inference pipeline to models/credit_inference_pipeline.pkl")

# 15) Prediction helper (example usage)
def predict_new(records_df, model_path="models/credit_inference_pipeline.pkl", proba_threshold=0.5):
    """
    records_df: pandas DataFrame with same columns used for features before preprocessing
    returns: DataFrame with prob and predicted label (1=bad)
    """
    model = joblib.load(model_path)
    probs = model.predict_proba(records_df)[:,1]
    preds = (probs >= proba_threshold).astype(int)
    out = records_df.copy()
    out["pred_proba_bad"] = probs
    out["pred_bad"] = preds
    return out

# Demo: predict on first 5 rows of X_test (before pipeline transform)
demo_in = X_test.iloc[:5].copy()
print("\nDemo predictions:")
display(predict_new(demo_in).T)

# 16) Save a short results summary to disk
summary = {
    "best_model": best_model_name,
    "test_roc_auc": float(roc_auc),
    "test_ap": float(ap),
    "test_precision": float(precision),
    "test_recall": float(recall),
    "test_f1": float(f1)
}
pd.Series(summary).to_frame("value").to_csv("models/summary_metrics.csv")
print("\nSaved summary metrics to models/summary_metrics.csv")
```
